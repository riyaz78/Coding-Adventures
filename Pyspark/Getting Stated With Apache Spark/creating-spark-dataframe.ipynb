## Reading Data using csv

```python
from pyspark.sql import SparkSession

spark  = SparkSession.builder \
            .appName("Getting Started With Sparks") \
            .getOrCreate()


read_df = spark.read \
            .format("csv") \
            .option("header","true") \
            .option("inferschema","true") \
            .load(r"C:\Users\rahmeds\source\My-Learning-Portfolio\Coding-Adventures\Pyspark\Sample Data\AI_Companies.csv")

read_df.show(20)
```

```python
import pandas as pd
import pandasql as ps


pandas_df = read_df.toPandas()

def sql_query(query):
    return ps.sqldf(query, {'pandas_df': pandas_df})

query = "SELECT * FROM pandas_df WHERE Location = 'Gurugram, India'"
result = sql_query(query)
display(result)

```
